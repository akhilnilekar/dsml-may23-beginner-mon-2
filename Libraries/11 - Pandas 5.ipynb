```python
def temp_mean(x):
    # x will be a group
    # in that group, we are setting the value of the entire Temp_Avg column
    # to the average value of the Temperature column
    
    x["Temp_Avg"] = x["Temperature"].mean()
    return x
```

```python
data_tidy.groupby(["Drug_Name"]).apply(temp_mean)
```

```python
data_final = data_tidy.groupby(["Drug_Name"], group_keys = False).apply(temp_mean)
```

```python
data_final
```

```python
data_final["Temperature"].fillna(data_final["Temp_Avg"], inplace = True)
```

```python
data_final["Temperature"].isna().sum()
```

```python
# STEP 1
def pressure_mean(x):
    x["Pressure_Avg"] = x["Pressure"].mean()
    return x
```

```python
# STEP 2
data_final = data_final.groupby(["Drug_Name"], group_keys = False).apply(pressure_mean)
```

```python
# STEP 3
data_final["Pressure"].fillna(data_final["Pressure_Avg"], inplace = True)
```

```python
data_final["Pressure"].isna().sum()
```

```python
temp_points = [5, 20, 35, 50, 60]
temp_labels = ["low", "medium", "high", "very high"]
```

```python
data_final["Temp_Category"] = pd.cut(
    data_final["Temperature"],
    bins = temp_points,
    labels = temp_labels
)
```
